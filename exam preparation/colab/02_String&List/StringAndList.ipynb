**Slicing Abbreviation**

```python
s = "Hello there, General Kenobi."
print(s[:5:])   #start at 0, to position 4, step 1 
```

```python
print(s[:5]) 
```

```python
x = [1,3,5,7,9,11]
print(x[1::2])  #start at 1, to last data, step 2
```

```python
print(x[:2:-1]) #not specify start but go back, 
                #will start at the last data
                #and stop before reaching position 2
```

```python
print(x[-2::-1])
```

```python
print(x[::-1])  #reverse list
```

**String Splitting (into list)**



```python
s = "Hello World    Oh my   gawd"
x = s.split()    #use white space to separate
print(x)
```

```python
s2 = "11:2:   33"
x = s2.split(":")  #specifying a separator
print(x)
```

```python
s3 = "a,,,b"
x = s3.split(",")
print(x)
```

```python
s4 = "a,,,b"
x = s4.split(",,")
print(x)
```

```python
s5 = "a   b"
x = s5.split(" ")
print(x)
```

```python
#6420008921,10,20,30   score for 3 quizzes
# Let's find the average.

s = "6420008921,10,20,30"
x = s.split(",")
avg = (int(x[1])+int(x[2])+int(x[3]))/3
print(avg)
```

```python
#changing input "day/month/Buddhistyear" to "day/month/AD"
x = input().split("/")
buddhistYear = int(x[2])
christianYear = buddhistYear-543
print(x[0]+"/"+x[1]+"/"+str(christianYear))
```
