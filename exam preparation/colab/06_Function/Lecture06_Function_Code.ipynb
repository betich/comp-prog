### Ex4: Compute a triangle area

Write a function <font color="red">triangle_area(a, b, c)</font> that gets the lengths of three triangle sides and calculate the area; $area=\sqrt{s(s-a)(s-b)(s-c)}$, where $s = \frac{a+b+c}{2}$

```python
# fill your code here
def triangle_area(a,b,c):
  s = (a+b+c)/2
  area = (s*(s-a)*(s-b)*(s-c))**1/2
  return area

print(triangle_area(2,3,4))
```

### Ex5: Prime number

Write a function <font color="red">is_prime(n)</font> to check as to whether or not $n$ (a positive integer) is prime.

```python
# fill your code here
def is_prime(n):
    # prime is the number that only divisible by itself and 1
    for i in range(2,n):
      if n%i != 0: 
        return False
    return True

is_prime(5)
```

### Ex6: Add two lists

Write a function <font color="red">add</font> to perform an element-wise add between two lists

```python
# fill your code here
def add(l1, l2):
  z = [0]* len(l1)
  for i in range(len(l1)):
    z[i] = l1[i] + l2[i]
  return z

x = [1, 2, 3]
y = [4, 5, 6]
# expected result => [1+4, 2+5, 3+6]
print(add(x,y))

```

Write a function <font color="red">fill(x,e)</font> to fill all values in a list $x$ with $e$, e.g., enter x=[0,1,2,3] & e=5 and the result is [5,5,5,5].

### Ex11: Replace values

Write a function <font color="red">replace(x, old, new, count)</font> to a relace an $old$ value with a $new$ value in a list $x$ with for $count$ times, e.g., call replace(x,3,9,2), where x=[1,2,3,3,3,2,3] and the result is [1,2,9,9,3,2,3]. 

Note that the result is updated to a given input and you shouldn't use 'return'.

```python
# fill code here
def replace(x, old, new, count):
  c = 0
  for i in range(len(x)):
    if x[i] == old and c<count:
      x[i] = new
      c+=1

x = [1,2,3,3,2,3]
print(x)
replace(x, 3, 9, 2)
print(x)
```
