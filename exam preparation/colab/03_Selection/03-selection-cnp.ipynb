## Boolean expression

The <it>condition</it> is a boolean expression which evaluates to either True or False.
<br>
Boolean expression can be written by using comparison operators which are:
+ `==` equal
+ `!=` not equal
+ `>` greater than
+ `<` less than
+ `>=` greater than or equal to
+ `<=` less than or equal to


````python
# True, False :boolean
print('1 == 1:', 1 == 1)
print('1 == 0:', 1 == 0)
print('0 == "0":', 0 == "0")
print((0.1+0.1+0.1) - 0.3 < 0.0000001)  
print(0.1+0.1+0.1, 0.3)
print(123456789123456789 == 123456789123456789)
````

````python
# logical expression == != > < >= <= 
# yields True or False
# try changing the values of a and b and see the result
a = 2
b = 3
print('{} == {} \t: {}'.format(a, b, a == b))
print('{} != {} \t: {}'.format(a, b, a != b))
print('{} > {} \t: {}'.format(a, b, a > b))
print('{} < {} \t: {}'.format(a, b, a < b))
print('{} >= {} \t: {}'.format(a, b, a >= b))
print('{} <= {} \t: {}'.format(a, b, a <= b))
````

## Rational operators (and, or, not)

combine logical expression

Let A, B are logical expressions
* A and B -> True if both are True, otherwise -> False
* A or B -> True if any is True. If both are False, -> False
* not True -> False, not False -> True

````python
# a and b
# not (a and b) -> not a or not b
# not (a or b) -> not a and not b 
x = 3
a = x > 5 # False
b = x < 10  # True
y = a and b # False
not_y = not a and not b  
#       True   and  False => False
print(a, b, y, not_y)
````

## Nested if
```python
if condition1:
  if condition2:
    # do this when both condition1 and condition2 are True

```
is equivalent to 
```python
if condition1 and condition2:
  # do this when both are True
```
The latter make the code easier to read.

````python
# nested if
x = 5
y = 4
z = 3

if x > y and y > z:
  print('x y z')

if x > y:
  if y > z:
    print('x > y > z')

if x > y > z:
  print('x > y > z')

if y < x and y < z:
  print('y is smallest')

print('-----------')
if x > y:
  a, b = 3, 4   # b 4 / a 3
  if x > z:
    a, b = -b, -a # a -4, b -3
  else:
    a, b = -a, -b
print(a, b)
````

## String comparison

Compare each character from left to right.
+ lowercase > UPPERCASE
+ alphabets have increasing value from `'A'` to `'Z'`
+ digits have increasing value from `'0'` to `'9'`
+ alphabet > digit

Try changing the value of s and t to see the comparison result

````python
# String comparison
print("'a' < 'b' :", 'a' < 'b')   # True
print("'A' < 'Z' :", 'A' < 'Z')   # True
print("'A' < 'a' :", 'A' < 'a')   # True
print("'a' < 'A' :", 'a' < 'A')   # False
print("'0' < 'A' :", '0' < 'A')   # True 
print("'0' < 'Z' :", '0' < 'Z')   # True

print('------------')
s = 'abc'
t = "abx"
print("'{}' == '{}' : {}".format(s, t, s == t )) # False
print("'{}' > '{}' : {}".format(s, t, s > t ))   # True
print("'{}' < '{}' : {}".format(s, t, s < t ))   # False
````

### Buy 2 get 1 free

````python
p1 = float(input('Price for item #1 : '))
p2 = float(input('Price for item #2 : '))
p3 = float(input('Price for item #3 : '))

min_p = p1
if p2 < min_p:
    min_p = p2
if p3 < min_p:
    min_p = p3

total_price = p1 + p2 + p3
print('Total:', total_price)
print('Got one free:', min_p)
print('Pay only:', total_price - min_p)
````
