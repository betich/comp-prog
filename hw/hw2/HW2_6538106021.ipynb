1. Write a function gradeNumber(x). This function takes a string letter grade (A,B+,B,C+,C,D+,D,F) and convert it to a corresponding numeric representation. For example, gradeNumber("C+") must return 2.5



2. Write function operate(filename). This function reads a file "filename" and calculates gpa and gpax for each semester. It returns a list containing lists of each semester's detail. For example, the sample "grades.txt" file will return the following list:

[['1st semester 2015', 2.25, 2.25], ['2nd semester 2015', 2.71, 2.44]] 

for each data x in a list, x[0] is the semester name, x[1] is the GPA for that semester, x[2] is a GPAX so far (including all previous semesters and the current semester). 


3. write function generateReport(list,filename). This function receives a list from part 2 and returns a string showing each semester's detail. It also writes this string into file filename.
  

For example, if our list parameter is [['1st semester 2015', 2.25, 2.25], ['2nd semester 2015', 2.71, 2.44]] , and the file name is "result1.txt" we will get a string :

And the file "result1.txt" will be created and will have this string written into the file. 


```python
#HW2 do not delete this line.
def gradeNumber(x): #write code for question 1
  if x == "F":
    return 0
  elif x == "D":
    return 1
  elif x == "D+":
    return 1.5
  elif x == "C":
    return 2
  elif x == "C+":
    return 2.5
  elif x == "B":
    return 3
  elif x == "B+":
    return 3.5
  elif x == "A":
    return 4
  else:
    return None 
    
def operate(filename): #write code for question 2
  f = open(filename, "r")
  grades = []
  
  for line in f:
    cols = line.split(",")
    if "semester" in line:
      grades.append([line.strip()])
    else:
      try:
        float(cols[2])
      except ValueError:
        continue
      else:
        grades[-1].append((gradeNumber(cols[3].strip()) * float(cols[2]), float(cols[2])))

  output = []

  get_gpa = lambda e: round(sum([x[0] for x in e]) / sum([x[1] for x in e]), 2)
  flatten = lambda l: [item for sublist in l for item in sublist[1:]]

  for i, e in enumerate(grades):
    output.append([e[0], get_gpa(e[1:]), get_gpa(flatten(grades[:i+1]))])

  return output
    

def generateReport(l,filename): #write code for question 3
  with open(filename, "w") as f:
    for e in l:
      f.write("{}\nGPA = {}, GPAX = {}\n".format(e[0], e[1], e[2]))
  
  return open(filename, "r").read()
    
    

#Test code for question 1
def test(expected, actual):
    print('Expected:\t', expected)
    print('Your result:\t', actual)

test(4,gradeNumber("A"))
test(3.5,gradeNumber("B+"))
test(3,gradeNumber("B"))
test(2.5,gradeNumber("C+"))
test(2,gradeNumber("C"))
test(1.5,gradeNumber("D+"))
test(1,gradeNumber("D"))
test(0,gradeNumber("F"))

#Test code for this question 2 
def test(expected, actual):
    print('Expected:\t', expected)
    print('Your result:\t', actual)

test([['1st semester 2015', 2.25, 2.25], ['2nd semester 2015', 2.71, 2.44]],operate("grades.txt"))
print("--------------")

expect = [['1st semester 2015', 2.65, 2.65], ['2nd semester 2015', 2.79, 2.71], ['1st semester 2016', 2.43, 2.62]]
test(expect,operate("grades2.txt"))

#Test code for question 3 (not including writing to file)
def test(expected, actual):
    print('Expected:\n', expected)
    print('Your result:\n', actual)
    
inputList = [['1st semester 2015', 2.25, 2.25], ['2nd semester 2015', 2.71, 2.44]]
expect = '1st semester 2015\nGPA = 2.25, GPAX = 2.25\n2nd semester 2015\nGPA = 2.71, GPAX = 2.44\n'
test(expect,generateReport(inputList,"result1.txt"))
print("---------------")

inputList = [['1st semester 2015', 2.65, 2.65], ['2nd semester 2015', 2.79, 2.71], ['1st semester 2016', 2.43, 2.62]]
expect = '1st semester 2015\nGPA = 2.65, GPAX = 2.65\n2nd semester 2015\nGPA = 2.79, GPAX = 2.71\n1st semester 2016\nGPA = 2.43, GPAX = 2.62\n'
test(expect,generateReport(inputList,"result2.txt"))
print("--------------")


#main program code is provided for you 
# so that you can see the overall operation when everything is put together.
gpalist = operate("grades.txt")
r = generateReport(gpalist,"finalResult1.txt")
print(r)

gpalist = operate("grades2.txt")
r = generateReport(gpalist,"finalResult2.txt")
print(r)




```
