```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#1

def generate_adjacency_matrix(filename):
  file = open(filename,"r")
  lines = file.readlines()
  person_names = set()
  friends = dict()
  for line in lines:
    line = line.strip().split(',')
    person_names.add(line[0])
    #print(line)
  person_names = sorted(person_names)
  A = np.zeros((len(person_names),len(person_names)),int)
  for line in lines:
    name,friend = line.strip().split(',')  
    a,b = person_names.index(name),person_names.index(friend)
    A[a,b] = 1
  file.close()



  
  return A, person_names
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#2

def get_degree_matrix(A):
  # write your code here
  D = np.zeros((len(A),len(A)),int)
  S = np.sum(A,axis = 1)
  for i in range(5):
    D[i,i] = S[i]



  return D
```

```python
D = get_degree_matrix(A)
print(D)
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#3

def get_names_with_highest_number_of_friends(D, person_names):
  # write your code here
  all_max = list()
  max_value = max(np.max(D,axis = 1))
  index_max_value = np.argmax(D,axis = 1)
  for i in index_max_value:
    if D[i,i] == max_value:
      all_max.append((person_names[i],i))
      
  return all_max
print(get_names_with_highest_number_of_friends(D, person_names))
```

```python
persons = get_names_with_highest_number_of_friends(D, person_names)
print(persons)
```

```python
# HELPER FUNCTION
def get_liked_page_matrix(infile):
  liked_pages = []
  persons = []
  fr = open(infile, 'r')
  for line in fr:
    name, bits = line.strip().split(',')
    liked_page = [int(c) for c in bits]
    liked_pages.append(liked_page)
    persons.append(name)
  liked_pages_array = np.array(liked_pages)
  return liked_pages_array,persons
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#4

def get_distance_matrix(Lp):
  # write your code here
  distance_matrix = np.zeros((len(Lp),len(Lp)),float)
  rows = list(Lp)
  data = list()
  for j in rows:
    data.append(list(j))
  for i in range(1,Lp.shape[0]): 
    A = Lp[i-1]
    for k in range(Lp.shape[0]):
      B = Lp[k]
      a = np.sum(A)
      b = np.sum(B)
      c = np.sum( (A == B) & ( A != 0) )
      cof = c/(a+b-c)
      if data.index(list(A)) == data.index(list(B)):
        pass
      else:
        distance_matrix[data.index(list(A)),data.index(list(B))] = cof
        distance_matrix[data.index(list(B)),data.index(list(A))] = cof


  return distance_matrix
```

```python
Dt = get_distance_matrix(liked_pages)
print(Dt)
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#5

def get_all_most_similar_pairs(Dt, persons):
  # write your code here
  max = np.max(Dt)
  pairs = list()
  for i in range(len(Dt)):
    for j in range(i+1,len(Dt)):
      if Dt[i,j] == max:
        pairs.append((persons[i],persons[j]))
      
      



  return pairs
```

```python
pairs = get_all_most_similar_pairs(Dt, persons)
print(pairs)
```
