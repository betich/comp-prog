```python
import numpy as np
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#1

def generate_adjacency_matrix(filename):
  # write your code here
  fn = open(filename,"r")
  A = [line.strip().split(",") for line in fn]
  fn.close()
  person_names = []
  person_names = [k for i in A for k in i if k not in person_names]
  person_names = set(person_names); person_names = list(person_names)
  person_names.sort()
  C = np.zeros((len(person_names),len(person_names)),int)
  for i in person_names:
      for e in A:
          if e[0] in i:
              C[person_names.index(e[0]), person_names.index(e[1])] = 1
  A = C
  return A, person_names
```

```python

# from google.colab import files
# uploaded = files.upload()
# print('uploaded = ', uploaded)

A, person_names = generate_adjacency_matrix("social_network.txt")
print(A)
print(person_names)
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#2

def get_degree_matrix(A):
  # write your code here
  B = np.sum(A,axis = 1)
  D = np.identity(len(A),int)
  D = B*D
  
  return D
```

```python
D = get_degree_matrix(A)
print(D)
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#3

def get_names_with_highest_number_of_friends(D, person_names):
  # write your code here
  A = D == np.max(D)
  A = np.argwhere(A == True)
  all_max = [(person_names[i[0]],i[0]) for i in A]
  return all_max
```

```python
persons = get_names_with_highest_number_of_friends(D, person_names)
print(persons)
```

```python
# HELPER FUNCTION
def get_liked_page_matrix(infile):
  liked_pages = []
  persons = []
  fr = open(infile, 'r')
  for line in fr:
    name, bits = line.strip().split(',')
    liked_page = [int(c) for c in bits]
    liked_pages.append(liked_page)
    persons.append(name)
  liked_pages_array = np.array(liked_pages)
  return liked_pages_array,persons
```

```python

# from google.colab import files
# uploaded = files.upload()
# print('uploaded = ', uploaded)

liked_pages, persons = get_liked_page_matrix('liked_pages.txt')
print(liked_pages)
print(persons)
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#4

def get_distance_matrix(Lp):
  # write your code here
  A = np.sum(Lp, axis=1)
  C = [sum(Lp[i,][(Lp[i, ::] == 1) & (Lp[e, ::] == 1)]) for i in range(Lp.shape[0]) for e in range(Lp.shape[0])]
  C = np.array(C)
  C = C.reshape((Lp.shape[0], Lp.shape[0]))
  np.fill_diagonal(C, 0)
  distance_matrix= [C[i][e] / (A[i] + A[e] - C[i][e])for i in range(A.shape[0]) for e in range(A.shape[0])]
  distance_matrix = np.array(distance_matrix)
  distance_matrix = distance_matrix.reshape((Lp.shape[0], Lp.shape[0]))

  return distance_matrix
```

```python
Dt = get_distance_matrix(liked_pages)
print(Dt)
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#5

def get_all_most_similar_pairs(Dt, persons):
  # write your code here
  A = Dt == np.max(Dt)
  A = np.argwhere(A == True)
  A = A.tolist()
  pairs = []
  for i in A:
      i = sorted(i)
      if (persons[i[0]],persons[i[1]]) not in pairs:
          pairs.append((persons[i[0]],persons[i[1]]))

  return pairs
```

```python
pairs = get_all_most_similar_pairs(Dt, persons)
print(pairs)
```
